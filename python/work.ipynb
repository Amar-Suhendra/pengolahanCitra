```python
import cv2 as cv
import matplotlib.pyplot as plt

# Read image
img = plt.imread('images/wtc.jpg')
hist = cv.calcHist([img], [0], None, [256], [0, 256])

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
uqualized = cv.equalizeHist(gray)

# Plot histogram and show image side by side 2x2
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs[0, 0].imshow(gray, cmap='gray')
axs[0, 0].set_title('Original')
axs[0, 1].imshow(uqualized, cmap='gray')
axs[0, 1].set_title('Equalized')
axs[1, 0].hist(gray.ravel(), 256, [0, 256])
axs[1, 0].set_title('Original Histogram')
axs[1, 1].hist(uqualized.ravel(), 256, [0, 256])
axs[1, 1].set_title('Equalized Histogram')
plt.show()
```
